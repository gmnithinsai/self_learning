# HashTable Implementation using Python
---

```python
class HashTable:
    def __init__(self):
        self.MAX = 100
        self.arr = [[] for i in range(self.MAX)]
        
    def get_hash(self, key):
        h = 0
        for char in key:
            h+=ord(char)
        return h%100
    
    def __setitem__(self, key, value):
        h_value = self.get_hash(key)
        found = False
        for idx, element in enumerate(self.arr[h_value]):
            if len(element) == 2 and element[0] == key:
                self.arr[h_value][idx] = (key, value)
                found = True
                break
        if not found:
            self.arr[h_value].append((key, value))
            
    def __getitem__(self, key):
        h_value = self.get_hash(key)
        for element in self.arr[h_value]:
            if element[0] == key:
                return element[1]
            
    def __delitem__(self, key):
        h_value = self.get_hash(key)
        for idx, element in enumerate(self.arr[h_value]):
            if element[0] == key:
                del self.arr[h_value][idx]
        
        

ht = HashTable()
        
```

```python
ht['march 17'] = 45
ht['march 6'] = 342
ht['dec 6'] = 67
ht['may 2'] = 10
ht['feb 10'] = 23
```
